```python
import pandas as pd
```

```python
wichita = pd.read_csv("https://datajournalism.tech/wp-content/uploads/2019/10/wichita.csv")
```

```python
wichita
```

```python
four = wichita[wichita.subject_race != "other/unknown"]
```

```python
populations = pd.DataFrame({"subject_race":["white", "black", "asian/pacific islander", "hispanic"],
              "num_people":[246343, 42679, 19272, 63659]},
            columns = ["subject_race", "num_people"])
```

```python
four.head()
```

```python
stops = pd.merge(four, populations, on = "subject_race")
```

```python
stops.head()
```

```python
four.subject_race.value_counts()
```

```python
four.head()
```

```python
race = four.subject_race.value_counts().reset_index()
```

```python
race = race.rename(columns={"index":"subject_race", "subject_race":"count"})
```

```python
stops = pd.merge(populations, race, on = "subject_race")
```

```python
stops
```

```python
stops["stop_rate"] = stops["count"]/stops["num_people"] * 100
```

```python
stops = stops.sort_values("stop_rate", ascending=False)
```

```python
stops
```

```python
%matplotlib inline
```

```python
stops.stop_rate.plot.barh()
```

```python
chart = stops.stop_rate.plot.barh()
```

```python
chart.set_yticklabels(stops.subject_race)
```

```python
chart
```

```python
stops.to_csv("stops.csv")
```
